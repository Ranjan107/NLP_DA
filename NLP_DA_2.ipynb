```python
import nltk
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
docs = """I am Ranjan. I am doing my enginnering at VIT Chennai. I am interested in Natural Language Processing. My hobbies are 
watching cricket, football, badminton. The sports I play are chess, cricket, badminton, table tennis etc. I have done many 
projects. I am interested in coding and working on projects. My project in natural language processing is sentimental analysis.  
Sentiment analysis is contextual mining of text which identifies and extracts subjective information in source material, and 
helping a business to understand the social sentiment of their brand, product or service while monitoring online conversations. 
However, analysis of social media streams is usually restricted to just basic sentiment analysis and count based metrics. This is 
akin to just scratching the surface and missing out on those high value insights that are waiting to be discovered. Sentiment
Analysis is the most common text classification tool that analyses an incoming message and tells whether the underlying sentiment
is positive, negative or neutral. I have also worked on many other projects such as movie revenue prediction using Machine Learning
, smart water quality predicting system using Artificial Intelligence, smart traffic management system using Internet of Things.
Hence, this is the corpus for digital assignment two.
"""
```

# Word Segmentation

```python
from wordsegment import load, segment
load()
```

```python
segment(docs)
```

# Sentence Segmentation

```python
sentences = nltk.sent_tokenize(docs)
```

```python
for sentence in sentences:
    print(sentence)
```

# Tokenization

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
tokenized_sent = tokenizer.tokenize(docs)
tokenized_sent
```

```python
len(tokenized_sent)
```

# Convert to lowercase

```python
words = []
for word in tokenized_sent:
    words.append(word.lower())
    
words[:19]
```

# Stop Words Removal

```python
processes_text = [word for word in words if word not in stopwords.words('english')]
processes_text
```

# Stemming

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
for w in processes_text:
    print(ps.stem(w))
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer()
lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in processes_text])
print(lemmatized_output)
```

# POS Tagger

```python
nltk.pos_tag(processes_text)
```
